### GroupBy

```python
# Importação de bibliotecas do PySpark
import pyspark.sql.functions as func
#from pyspark.sql import DataFrameStatFunctions as statFunc
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, when
from itertools import groupby
from pandas.core.groupby.groupby import GroupBy
```

```python
# Cálculo da média das taxas de aprovação nos anos iniciais do ensino fundamental por unidade geográfica com NaN
df_fund_aprov_geo_mean_nan = df_py.groupBy('und_geografica').agg(func.mean(df_py['anos_iniciais_fund_aprov']).
                                                          alias('media_aprovacao_anos_iniciais_fundamental'))
df_fund_aprov_geo_mean_nan.show()
```

```python
# Colocando nas medidas estatísticas adicionais, como variância, quartis, limite superior e limite inferior
sd = list(medidas.iloc[2])            # sd recebe uma lista com o conteúdo de índice 2 do df medidas
sd.pop(0)                             # Removendo o item de índice 0
q1 = list(medidas.iloc[4])            # q1 recebe uma lista com o conteúdo de índice 4 do df medidas
q1.pop(0)                             # Removendo o item de índice 0
q3 = list(medidas.iloc[6])            # # q3 recebe uma lista com o conteúdo de índice 6 do df medidas
q3.pop(0)                             # Removendo o item de índice 0
min = list(medidas.iloc[3])           # min recebe uma lista com o conteúdo de índice 3 do df medidas
min.pop(0)                            # Removendo o item de índice 0
max = list(medidas.iloc[7])           # max recebe uma lista com o conteúdo de índice 7 do df medidas
max.pop(0)                            # Removendo o item de índice 0
var = []                        # Criação de lista vazia var para variância
at = []                         # Criação de lista vazia at para amplitude total
aiq = []                        # Criação de lista vazia aiq para amplitude interquartílica
ls = []                         # Criação de lista vazia ls para limite superior
li = []                         # Criação de lista vazia li para limite inferior
for i in range(len(sd)):              # Percorrendo o comprimento de sd
  var.append(sd[i]**2)                # Colocando em var sd**2
  at.append(max[i]-min[i])            # Colocando em at o resultado de max - min
  aiq.append(q3[i]-q1[i])             # Colocando em aiq o resultado de q3-q1
  ls.append(q3[i]+(1.5*aiq[i]))       # Colocando em ls o resultado de q3 + 1.5*aiq
  li.append(q1[i]-(1.5*aiq[i]))       # Colocando em li o resultado de q1 - 1.5*aiq
var.insert(0, 'var')            # Inserindo a variância 'var' na posição 0
at.insert(0, 'AT')                    # Inserindo 'AT' na posição 0
aiq.insert(0, 'AIQ')                  # Inserindo 'AIQ' na posição 0
ls.insert(0, 'LS')                    # Inserindo 'LS' na posição 0
li.insert(0, 'LI')                    # Inserindo 'LI' na posição 0
medidas.loc[9] = var            # Colocando var na lina 9
medidas.loc[10] = at            # Colocando at na lina 10
medidas.loc[11] = aiq           # Colocando aiq na lina 11
medidas.loc[12] = ls            # Colocando ls na lina 12
medidas.loc[13] = li            # Colocando li na lina 13
display(medidas)                      # Mostrando os novos dados de medidas
```
