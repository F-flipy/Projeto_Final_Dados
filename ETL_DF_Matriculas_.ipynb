## DF Idades

- Ano - ano: Ano de realização do censo;
- Unidade Geográfica - unid_geográfica: Região e Estados com dados do censo;
- 6_10_anos : Soma dos alunos matriculados de 6 a 10 anos;
- 11_14_anos : Soma dos alunos matriculados de 11 a 14 anos;
- 15_17_anos : Soma dos alunos matriculados de 15 a 17 anos;

## DF  Sexo e Cor
- Ano - ano: Ano de realização do censo;
- und_geografica: Região e Estados com dados do censo;
- total_fem : total femininos
- nao_dec_fem : Não declarado feminino
- branca_fem : mulheres brancas
- preta_fem : mulheres pretas
- parda_fem : mulheres pardas
- amarela_fem	: mulheres amarelas
- indígena_fem : mulheres indigenas
- total_masc : total masculino
- nao_dec_masc : Não declarado masculino
- branca_masc : homens brancos
- preta_masc : homens pretos
- parda_masc	: homens pardos
- amarela_masc : homens amarelos
- indígena_masc : homens indigenas



## DF Taxa de rendimento

- Ano - ano: Ano de realização do censo;
- Unidade Geográfica - unid_geográfica: Região e Estados com dados do censo;
- Localização - localizacao : Matriculas realizadas de acordo com a localização zona rural ou urbana;
- Dependencia Administrativa - dependencia_adm : Matrículas realizadas de acordo com o tipo de escola Estadual, Municipal,
- Pública, Federal ou Privada;
- Total - total_fund_aprov : Número total de alunos aprovados no Ensino Fundamental;
- Anos Iniciais - anos_finais_fund_aprov : Número total de alunos aprovados nos Anos Iniciais do Ensino Fundamental, inclui turmas do 1º ao 5º ano;
- Anos Finais - anos_iniciais_fund_aprov : Número total de alunos aprovados nos Anos Iniciais do Ensino Fundamental, inclui turmas do 6º ao 9º ano;
- total_medio_aprov : Número total de alunos aprovados no Ensino Médio;
- total_fund_reprov : Número total de alunos reprovados no Ensino Fundamental;
- anos_finais_fund_reprov : Número total de alunos reprovados nos Anos Iniciais do Ensino Fundamental, inclui turmas do 1º ao 5º ano;
- anos_iniciais_fund_reprov : Número total de alunos reprovados nos Anos Iniciais do Ensino Fundamental, inclui turmas do 6º ao 9º ano;
- total_medio_reprov : Número total de alunos reprovados no Ensino Médio;
- total_fund_aband : Número total de alunos que abandonaram o Ensino Fundamental;
- anos_finais_fund_aband : Número total de alunos que abandonaram o Ensino Fundamental, nos Anos Iniciais, inclui turmas do 1º ao 5º ano;
- anos_iniciais_fund_aband : Número total de alunos que abandonaram o Ensino Fundamental, nos Anos Iniciais, inclui turmas do 6º ao 9º ano;
- total_medio_aband : Número total de alunos que abandonaram o Ensino Médio.

## DF Matriculas

- ano: Ano de realização do censo;
- unid_geográfica: Região e Estados com dados do censo;
total_fund : Soma total das matrículas do Ensino Fundamental, inclui turmas do 1º ao 9º ano;
- anos_iniciais : Soma dos Anos Iniciais do Ensino Fundamental, inclui matrículas em turmas do 1º ao 5º ano;
- anos_finais : Soma dos Anos Iniciais do Ensino Fundamental inclui matrículas em turmas do 6º ao 9º ano;
- total_medio : Soma total do Ensino Médio inclui matrículas do Ensino Médio - Propedêutico, Curso Técnico Integrado (Ensino Médio Integrado) e Ensino Médio Normal/Magistério;
- total_geral : Soma total das matrículas realizadas em cada período (ano) de censo e região geográfica
